```python
with open('input/ctf.txt') as f:
    text = f.read()
```

```python
filtered_logs = [v for v in text.splitlines() if v.startswith('200')]

filtered_logs
```

```python
not_standard = [v for v in filtered_logs if "179B" not in v]

not_standard
```

```python
import requests

urls = [v[13:] for v in filtered_logs]
downloaded_urls = {}

for url in urls:
    if url.endswith('/'):
        continue
    
    r = requests.get(url, allow_redirects=True, headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"})
    downloaded_urls[url] = r.content
    
    if "flag" in r.text or "Flag" in r.text:
        print(r.content)
    
downloaded_urls
```
